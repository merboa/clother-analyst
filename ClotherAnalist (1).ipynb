```python
import numpy as np
import pandas as pd

from datetime import datetime

from matplotlib import pyplot as plt
import seaborn as sns
import glob
```

```python
clother_list = ['Aizel', 
                'Bosco', 
                'Markformelle', 
                'Pompa']               
```

```python
def make_combined_category_names():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_categories_names = clother+'_categories_names.csv'
        df_categories_names = pd.read_csv(file_name_categories_names)
        df_full = df_full.append(df_categories_names).assign(shop = clother )
        
        
    return df_full
```

```python
def make_combined_profuct_names():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_profucts = clother+'_products_names.csv'
        df_products_names = pd.read_csv(file_name_profucts)
        df_full = df_full.append(df_products_names)
        
    return df_full
```

```python
def make_combined_category():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_categories = clother+'_categories.csv'
        df_categories = pd.read_csv(file_name_categories)
        df_full = df_full.append(df_categories)
        
    return df_full
```

```python
def make_combined_profuct():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_products = clother+'_products.csv'
        df_products = pd.read_csv(file_name_products)
        df_full = df_full.append(df_products)
        
    return df_full
```

```python
def make_combined_purchases():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_purchases = clother+'_purchases.csv'
        df_purchases = pd.read_csv(file_name_purchases)
        df_full = df_full.append(df_purchases).assign(shop = clother )
        
    return df_full
```

```python
def make_combined_customers():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_customers = clother+'_customers.csv'
        df_customers = pd.read_csv(file_name_customers)
        df_full = df_full.append(df_customers)
        
    return df_full
```

```python
def make_combined_attribution():
    
    df_full = pd.DataFrame()

    for clother in clother_list:
        file_name_attribution = clother+'_attribution.csv'
        df_attribution = pd.read_csv(file_name_attribution)
        df_full = df_full.append(df_attribution).assign(shop = clother )
        
    return df_full
```

```python

df_products = make_combined_profuct()
df_purchases = make_combined_purchases()
df_customers = make_combined_customers()
```

```python
df_products_purchases = df_purchases.merge(df_products, on = 'product_id', how = 'left')
df_products_purchases = df_products_purchases [['datetime','customer_id','product_id','price', 'shop']]
df_products_purchases.head()
```

```python
df_products_purchases_full = df_products_purchases.merge(df_customers, on = 'customer_id', how = 'left')
df_products_purchases_full = df_products_purchases_full [['datetime','customer_id','product_id','price','sex','shop']]
df_products_purchases_full.head()
```

```python
df_products_purchases_full['datetime'] = pd.to_datetime(df_products_purchases_full['datetime'] )
```

```python
df_products_purchases_full['year'] = df_products_purchases_full['datetime'].dt.year
df_products_purchases_full['month'] = df_products_purchases_full['datetime'].dt.month
df_products_purchases_full['day'] = df_products_purchases_full['datetime'].dt.day
df_products_purchases_full['hour'] = df_products_purchases_full['datetime'].dt.hour
df_products_purchases_full
```

```python
df_products_purchases_full = df_products_purchases_full.set_index('datetime')
df_products_purchases_full
```

#### 1.1 All Purchases 2020 jan - aug

```python
df_selected_1 = df_products_purchases_full.loc['2020-01-01':'2020-08-31'][['customer_id', 'product_id', 'price','sex','hour','shop']]
df_selected_1
```

#### 2.1 All purchases 2021 (jan - aug)

```python
df_selected_2 = df_products_purchases_full.loc['2021-01-01':'2021-08-31'][['customer_id', 'product_id', 'price','sex','hour','shop']]
df_selected_2
```

```python
df_by_hour_1fem = df_selected_1\
    .loc[df_selected_1['sex'] == 'Female']\
    .groupby(['hour'])\
    .sum()[['price']]
```

```python
df_by_hour_1fem.to_excel('20fem.xlsx')
```

```python
df_by_hour_2fem = df_selected_2\
    .loc[df_selected_2['sex'] == 'Female']\
    .groupby(['hour'])\
    .sum()[['price']]
```

```python
df_by_hour_2fem.to_excel('21fem.xlsx')
```

#### 3.3 Order by hour (male)

```python
df_by_hour_1male = df_selected_1\
    .loc[df_selected_1['sex'] == 'Male']\
    .groupby(['hour'])\
    .sum()[['price']]
```

```python
df_by_hour_1male.to_excel('20mal.xlsx')
```

```python
df_by_hour_2male = df_selected_2\
    .loc[df_selected_2['sex'] == 'Male']\
    .groupby(['hour']).sum()[['price']]
```

```python
df_by_hour_2male.to_excel('21mal.xlsx')
```
